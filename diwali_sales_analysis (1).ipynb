### Age

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame and it's already defined
# df = pd.read_csv('your_data.csv')  # Example for loading data

# Grouping the data
age_gender_counts = df.groupby(['Age Group', 'Gender']).size().unstack(fill_value=0)

# Create a bar plot using Matplotlib
fig, ax = plt.subplots()

# Define colors for each gender (ensure the number of colors matches the number of hues)
colors = ['skyblue', 'pink']  # Example colors for different genders

# Define the width of each bar and position of bars
width = 0.4
x = range(len(age_gender_counts.index))

# Plot bars for each gender separately
for i, gender in enumerate(age_gender_counts.columns):
    bars = ax.bar(
        [pos + i * width for pos in x],  # Position bars side by side
        age_gender_counts[gender],
        width=width,
        color=colors[i],  # Use color for the current gender
        label=gender
    )

# Adding labels to the bars
for bar in ax.patches:
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # X position (center of the bar)
        height,  # Y position (top of the bar)
        f'{height}',  # Text label
        ha='center',  # Horizontal alignment
        va='bottom'  # Vertical alignment
    )

# Adding title and labels for better understanding
ax.set_title('Count of Age Groups by Gender')
ax.set_xlabel('Age Group')
ax.set_ylabel('Count')

# Set the x-ticks to be in the middle of the grouped bars
ax.set_xticks([pos + width / 2 * (len(age_gender_counts.columns) - 1) / 2 for pos in x])
ax.set_xticklabels(age_gender_counts.index)

# Adding legend
ax.legend(title='Gender')

plt.show()
```

```python
import pandas as pd

# Load the CSV file (make sure the filename is correct)
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the sales_state DataFrame
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum()

# Pivot the data to ensure proper alignment
sales_pivot = sales_state.pivot(index='Marital_Status', columns='Gender', values='Amount').fillna(0)

# Set the figure size
plt.figure(figsize=(6, 5))

# Define bar width and positions
bar_width = 0.4
positions = np.arange(len(sales_pivot.index))

# Iterate over each Gender to create bars
for i, gender in enumerate(sales_pivot.columns):
    plt.bar(positions + i * bar_width, sales_pivot[gender], width=bar_width, label=gender)

# Customize the x-axis labels
plt.xticks(positions + bar_width / 2, sales_pivot.index)

# Add labels and title
plt.xlabel('Marital Status')
plt.ylabel('Amount')
plt.title('Amount by Marital Status and Gender')

# Add a legend
plt.legend(title='Gender')

# Show the plot
plt.show()
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
import matplotlib.pyplot as plt
import numpy as np

# Count the occurrences of each unique value in the 'Occupation' column
occupation_counts = df['Occupation'].value_counts()

# Set the figure size
plt.figure(figsize=(20, 5))

# Get a colormap with a number of colors equal to the number of unique occupations
colors = plt.colormaps['tab20'](np.linspace(0, 1, len(occupation_counts)))

# Create the bar plot with different colors
bars = plt.bar(occupation_counts.index, occupation_counts.values, color=colors)

# Add labels to each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

# Add labels and title
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Count of Each Occupation')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Group by 'Product_ID', sum the 'Orders', get the top 10, and sort them
top_10_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)

# Set up the figure and axis
fig1, ax1 = plt.subplots(figsize=(12, 7))

# Define a colormap with enough colors
colors = plt.colormaps['tab20'](np.linspace(0, 1, len(top_10_products)))

# Create the bar plot with different colors
bars = ax1.bar(top_10_products.index, top_10_products.values, color=colors)

# Add labels to each bar
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

# Add labels and title
ax1.set_xlabel('Product ID')
ax1.set_ylabel('Orders')
ax1.set_title('Top 10 Most Sold Products')

# Rotate x-axis labels for better readability
ax1.set_xticklabels(top_10_products.index, rotation=45, ha='right')

# Display the plot
plt.show()
```
